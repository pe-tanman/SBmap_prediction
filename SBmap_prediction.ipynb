```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as mpl
import math

max_time = 7200
times = range(0, max_time)
altitudes = [0,4]
label = "a_n=sqrt(a_(n-1))+a_(n-2)"

def current_altitude(time):
    result = math.sqrt(altitudes[time-1]) + altitudes[time-2]
    return result

def plot(y_list):
    fig = mpl.figure()
    mpl.xlabel("x")
    mpl.ylabel("y")
    mpl.grid(True)
    mpl.scatter(times, y_list,s=1, label=label) 

    mpl.legend()
    mpl.show()
    fig.savefig(label+".png")

#main
for t in range(2, max_time):
    altitudes.append(current_altitude(t))

plot(altitudes)
```
